```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Загрузка данных
df = pd.read_csv('C:/Users/epg_F/clients.csv')
df.head(10)
```

```python
# Предварительный анализ
print(df.info(10))
print(df.describe())
```

```python
# 1. Пол целевого клиента
plt.figure(figsize=(5, 5))
sns.countplot(data=df, x='Gender')
plt.title('Распределение по полу клиентов')
plt.xlabel('Пол')
plt.ylabel('Количество клиентов')
plt.show()
```

```python
# 2. Возраст целевого клиента
plt.figure(figsize=(5, 5))
sns.histplot(df['Age Interval '], bins=10, kde=True)
plt.title('Распределение возрастов клиентов')
plt.xlabel('Возраст')
plt.ylabel('Количество клиентов')
plt.show()
```

```python
# 3. Тип квартиры
plt.figure(figsize=(5, 6))
sns.countplot(data=df, x='Type of property')
plt.title('Распределение типов квартир')
plt.xlabel('Тип квартиры')
plt.ylabel('Количество клиентов')
plt.xticks(rotation=45)
plt.show()
```

```python
# 4. Площадь квартиры
plt.figure(figsize=(5, 5))
sns.boxplot(data=df, y='Area (ft.)')
plt.title('Распределение площади квартир')
plt.ylabel('Площадь (кв. ф)')
plt.show()
```

```python
# Очистка столбца Price без обратного слэша
df['Price'] = df['Price'].replace({'$': '', ',': ''}, regex=True).astype(float)
average_price = df['Price'].mean()
print(f'Средний чек: ${average_price:.2f}')

plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], bins=30, kde=True, color='skyblue', edgecolor='black')
plt.axvline(average_price, color='red', linestyle='dashed', linewidth=2, label='Средний чек')
plt.title('Распределение цен')
plt.xlabel('Цена ($)')
plt.ylabel('Количество клиентов')
plt.legend()
plt.grid(axis='y')
plt.show()
```

```python
# 6. Цели покупки
plt.figure(figsize=(5, 5))
sns.countplot(data=df, x='Purpose')
plt.title('Цели покупки квартир')
plt.xlabel('Цель покупки')
plt.ylabel('Количество клиентов')
plt.xticks(rotation=45)
plt.show()
```

```python
# 7. Источник лида
plt.figure(figsize=(5, 6))
sns.countplot(data=df, x='Source')
plt.title('Источники лидов')
plt.xlabel('Источник лида')
plt.ylabel('Количество клиентов')
plt.xticks(rotation=45)
plt.show()
```
